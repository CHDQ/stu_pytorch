# FizzBuzz
游戏规则,从1开始往上数，遇到3的倍数，说fizz。遇到5的倍数，说buzz。遇到15的倍数说FizzBuzz.

```python
def fizz_buzz_encode(i):
    if i%15==0:
        return 3
    if i%5==0:
        return 2
    if i%3==0:
        return 1
    return 0
```

```python
 def fizz_buzz_decode(i):
    return [str(i),"Fizz","Buzz","FizzBuzz"][fizz_buzz_encode(i)]
```

```python
for i in range(20):
    print(fizz_buzz_decode(i))
```

# 机器学习实现
## 定义输入

```python
import numpy as np
import torch
```

```python
NUM_DIGITS=10
def encode_binary(i,num_digits):
    return np.array([i>>x&1 for x in range(num_digits)][::-1])
tr_x=torch.Tensor([encode_binary(x,NUM_DIGITS) for x in range(101,2**NUM_DIGITS)])
tr_y=torch.LongTensor([fizz_buzz_encode(x) for x in range(101,2**NUM_DIGITS)])
```

```python
tr_y.shape
```

```python
NUM_HIDDEN=100
model=torch.nn.Sequential(
    torch.nn.Linear(NUM_DIGITS,NUM_HIDDEN),
    torch.nn.ReLU(),
    torch.nn.Linear(NUM_HIDDEN,4)
).cuda()
```

```python
torch.cuda.is_available()
```

```python
loss_fn=torch.nn.CrossEntropyLoss()
optimizer=torch.optim.SGD(model.parameters(),lr=0.01)
```

```python
BATCH_SIZE=128
for i in range(10000):
    for start in range(0,len(tr_x),BATCH_SIZE):
        end=start+BATCH_SIZE
        batch_x=tr_x[start:end]
        batch_y=tr_y[start:end]
        batch_x=batch_x.cuda()
        batch_y=batch_y.cuda()
        y_pred=model(batch_x)
        loss=loss_fn(y_pred,batch_y)
        print("Epoch ",i,loss.item())
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
```

## 测试结果

```python
test_x=torch.Tensor([encode_binary(x,NUM_DIGITS) for x in range(1,101)])
with torch.no_grad():
    test_y=model(test_x.cuda())
pret=zip(range(1,101),test_y.cpu().data.max(1)[1].numpy())
print([(i,fizz_buzz_decode(v)) for i,v in pret])
```
